```python
BASE_URL = 'http://adslabs.org/adsabs/api/search/'
params = {'dev_key': '9UFkP4yT83I1uYWz',
 'fl': 'abstract,aff,keyword,title',
 'q': 'aff:France, keyword_schema:astronomy,year:2012',
 'rows': '200',
 'start': 0}
import requests
import json
```

```python
data = []
for i in range(2000) :
    params['start'] = i
    r = requests.get(BASE_URL, params=params)
    el_data = json.loads(r.text)
    data.append(el_data)
```

```python
for i in range(2000) :
    params['start'] = i
    params['q'] = 'aff:France, keyword_schema:astronomy,year:2013'
    r = requests.get(BASE_URL, params=params)
    el_data = json.loads(r.text)
    data.append(el_data)
```

```python
from pandas import *
flat_list = []
for n in range(len(data)) :
    for ndx in range(len(data[n]['results']['docs'])) :
        flat_list.append(data[n]['results']['docs'][ndx])
dtf = DataFrame(flat_list)
dtf
```

```python
import nltk
from nltk import *
```

```python
c_a_ppp = [len(i) for i in dtf['aff']]
```

```python
frdist = [FreqDist(word_tokenize(str(i))) for i in dtf['aff']]
```

```python

c_fr_pa = [i['France'] for i in frdist]
```

```python
dtf['count of authors'] = c_a_ppp
dtf['count of FR authors'] = c_fr_pa
dtf['percentage of FR authors'] = dtf['count of FR authors'] / dtf['count of authors']
```

```python
aff_b = [tuple(i) for i in dtf['aff']]
dtf['aff'] = aff_b
kw_b = [tuple(i) for i in dtf['keyword']]
dtf['keyword'] = kw_b
title_b = [tuple(i) for i in dtf['title']]
dtf['title'] = title_b
```
