```python
BASE_URL = 'http://adslabs.org/adsabs/api/search/'
params = {'dev_key': '9UFkP4yT83I1uYWz',
 'fl': 'abstract,aff,keyword,title,citation_count',
 'q': 'keyword_schema:astronomy, year:2010',
 'rows': '200',
 'start': 0}
import requests
import json
```

```python
data = []
for i in range(10) :
    params['start'] = i
    r = requests.get(BASE_URL, params=params)
    el_data = json.loads(r.text)
    data.append(el_data)
```

```python
from pandas import *
import numpy as np
some_list = [d['results']['docs'] for d in data]
some_array = np.array([i for i in some_list])
another_list = list(some_array.ravel())
dtf = DataFrame(another_list)
```
