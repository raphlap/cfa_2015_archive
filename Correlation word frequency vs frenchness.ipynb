```python
data = []
for i in range(10) :
    params['start'] = i
    r = requests.get(BASE_URL, params=params)
    el_data = json.loads(r.text)
    data.append(el_data)
```

```python
from pandas import *
import numpy as np
some_list = [d['results']['docs'] for d in data]
some_array = np.array([i for i in some_list])
another_list = list(some_array.ravel())
dtf = DataFrame(another_list)
```

```python
import nltk
from nltk import *
from collections import Counter
c_a_ppp = [len(i) for i in dtf['aff']]
frdist = [FreqDist(word_tokenize(str(i))) for i in dtf['aff']]
c_fr_pa = [i['France'] for i in frdist]
```

```python
dtf['count of authors'] = c_a_ppp
dtf['count of FR authors'] = c_fr_pa
dtf['percentage of FR authors'] = dtf['count of FR authors'] / dtf['count of authors']
```

```python
data_dict = dict(dtf)
f = open('/Users/lib-mac-107/Documents/NLTK_ADS_CORPUS/corr_sample.json', 'w')
f.write(json.dumps(data_dict))
f.close()
```

```python
DATA = '/Users/lib-mac-107/Documents/NLTK_ADS_CORPUS/corr_sample.json'

f = open(DATA).read()
Data = json.loads(f)
```

```python
abstracts = [word_tokenize(unicode(v)) for v in Data['abstract'].values()]
tc = nltk.TextCollection(abstracts)
```

```python
td_matrix = {}
for idx in range(len(abstracts)):
    abstract = abstracts[idx]
    fdist = nltk.FreqDist(abstract)

    bibcodes = Data['bibcode'][str(idx)]
    percentage_frenchness = Data['percentage of FR authors'][str(idx)]
    td_matrix[(bibcodes, percentage_frenchness)] = {}

    for term in fdist.iterkeys():
        td_matrix[(bibcodes, percentage_frenchness)][term] = tc.tf_idf(term, abstract)
```

```python
some_df = DataFrame(td_matrix)
percentages = list(some_df.columns.values)
list_percentages = [i[1] for i in percentages]
df = some_df.T
df['percentages'] = list_percentages
df = df.fillna(0)
```

```python
import warnings
warnings.simplefilter("error")
```

```python
dtf_corr = DataFrame(df.corrwith(df.percentages))
dtf_corr.sort_index(by=[0], ascending = True)
```
