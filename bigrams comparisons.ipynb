```python
import json
DATA = '/Users/lib-mac-107/Documents/NLTK_ADS_CORPUS/FR_2011.json'

f = open(DATA).read()
Data = json.loads(f)
```

```python
import nltk
from nltk import *
abstracts_fr = [word_tokenize(unicode(v)) for v in Data['abstract'].values()]
stop = nltk.corpus.stopwords.words('english')
filtered = []
for abstract in abstracts_fr :
    abstract = [w.lower() for w in abstract if w.isalpha() and w not in stop and len(w)>7]
    filtered.append(abstract)
filtered?
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
all_scores = []
for f in filtered :
    word_fd = nltk.FreqDist(f)
    bigram_fd = nltk.FreqDist(nltk.bigrams(f))
    finder = BigramCollocationFinder(word_fd, bigram_fd)
    finder.apply_word_filter(lambda w:w in nltk.corpus.stopwords.words('english'))
    scored = dict(finder.score_ngrams(bigram_measures.raw_freq))
    all_scores.append(scored)
all_scores
list_bi_fr = [list(i.keys()) for i in all_scores]
list_bi_fr
```

```python
bibi = {}
for idx in range(len(all_scores)):
    bibcodes = Data['bibcode'][str(idx)]
    bibi[(bibcodes)] = list_bi_fr[idx]
bibi
```
