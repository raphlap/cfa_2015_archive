```python
import json
import nltk
fi = open('/Users/lib-mac-107/Documents/analysis/citing_french.json').read()
Data = json.loads(fi)
```

```python
DATA = []
for i in range(len(Data)) :
    e = Data[i]['meta']['hits']
    if e != 0 :
        DATA.append(Data[i])
```

```python
from pandas import *
from nltk import *
```

```python
bibcode_citation_dict = {}
citing_bibcode_list = []
for i in range(len(DATA)) :
    dtf = DataFrame(DATA[i]['results']['docs']).dropna()
    dict_key= DATA[i]['meta']['query'][10:]
    c_a_ppp = [len(i) for i in dtf['aff']]
    frdist = [FreqDist(word_tokenize(str(i))) for i in dtf['aff']]
    c_fr_pa = [i['France'] for i in frdist]    
    dtf['count of authors'] = c_a_ppp
    dtf['count of FR authors'] = c_fr_pa        
    dtf['percentage of FR authors'] = dtf['count of FR authors'] / dtf['count of authors']
    bibcode_list = [b for b in dtf['bibcode']]
    percentage_list = [p for p in dtf['percentage of FR authors']]
    dict_citation = {}
    for i in range(len(bibcode_list)) :
        dict_citation[bibcode_list[i]] = percentage_list[i]
        citing_bibcode_list.append(tuple([bibcode_list[i], percentage_list[i]]))
    bibcode_citation_dict[dict_key] = dict_citation
```

```python
citing_bibcode_list = list(set(citing_bibcode_list))
```

```python
import json
DATA = '/Users/lib-mac-107/Documents/analysis/giant_french.json'
f = open(DATA).read()
Data = json.loads(f)
from pandas import * 
DTF = DataFrame(Data).reset_index(drop=True)
DTF[0:10]
```

```python
orig_bibcode_list = []
for i in range(len(DTF)) :
    orig_bibcode_list.append(tuple([DTF.ix[i]['bibcode'], DTF.ix[i]['percentage of FR authors']]))
```

```python
distance_df_list = citing_bibcode_list + orig_bibcode_list
distance_df_list = list(set(distance_df_list))
distance_df_list[0][0]
```

```python
def citing_indicator(bibcode1, bibcode2) :
    score = 0
    try :
        if bibcode2[0] in bibcode_citation_dict[bibcode1[0]].keys() :
            score += 1
        else :
            score = score
    except KeyError, e :
        score = score
    return score
```

```python
distances = {}
for (bibcode1, percentage1) in distance_df_list :
    distances[(bibcode1, percentage1)] = {}
    for (bibcode2, percentage2) in distance_df_list :
        distances[(bibcode1, percentage1)][(bibcode2, percentage2)] = citing_indicator((bibcode1, percentage1), (bibcode2, percentage2))
    
```

```python
df = DataFrame(distances)
df
```
