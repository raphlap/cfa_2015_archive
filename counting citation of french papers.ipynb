```python
from pandas import *
import json
DATA = '/Users/lib-mac-107/Documents/NLTK_ADS_CORPUS/FR_giant_data.json'
f = open(DATA).read()
data = json.loads(f)
data['abstract']['0']
DTF = DataFrame(data)
DTF = DTF.reset_index(drop=True)
```

```python
import numpy as np
threshold_list = np.arange(0,1,0.05)
threshold_list = [round(d, 2) for d in threshold_list]
```

```python
count_dict = {}
for i in threshold_list :
    filtered_df = DTF[DTF['percentage of FR authors'] >= i]
    query_list = ['reference:'+b for b in filtered_df['bibcode']]
    count_dict[i] = len(query_list)
```

```python
count = DataFrame(Series(count_dict)).reset_index()
count.columns = ['percentage of french authors', 'number of papers citing french papers']
count
```
