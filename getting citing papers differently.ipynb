```python
import json
import nltk
fi = open('/Users/lib-mac-107/Documents/analysis/citing_french.json').read()
Data = json.loads(fi)
```

```python
DATA = []
for i in range(len(Data)) :
    e = Data[i]['meta']['hits']
    if e != 0 :
        DATA.append(Data[i])
```

```python
DATA[0]['results']['docs'][0]['aff']
```

```python
import nltk
from nltk import *
```

```python
bibcode_citation_dict = {}
for i in range(len(DATA)) :
    dtf = DataFrame(DATA[i]['results']['docs']).dropna()
    dict_key= DATA[i]['meta']['query'][10:]
    c_a_ppp = [len(i) for i in dtf['aff']]
    frdist = [FreqDist(word_tokenize(str(i))) for i in dtf['aff']]
    c_fr_pa = [i['France'] for i in frdist]    
    dtf['count of authors'] = c_a_ppp
    dtf['count of FR authors'] = c_fr_pa        
    dtf['percentage of FR authors'] = dtf['count of FR authors'] / dtf['count of authors']
    mean = dtf['percentage of FR authors'].mean()
    bibcode_citation_dict[dict_key] = mean
  
```

```python
f = open('/Users/lib-mac-107/Documents/analysis/citing_french_citation_dict.json', 'w')
f.write(json.dumps(bibcode_citation_dict))
f.close()
```
